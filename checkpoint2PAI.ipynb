```python
# === Parte 4 — Montar o Modelfile (Versão Personal Trainer) ===

def gerar_modelfile(modelo_base, system_prompt, exemplos):
    linhas = []

    linhas.append(f"FROM {modelo_base}")
    linhas.append("PARAMETER temperature 0.7")
    linhas.append("PARAMETER num_predict 300")

    linhas.append(f'SYSTEM """{system_prompt}"""')

    for ex in exemplos:
        pergunta = ex.get("pergunta", "")
        resposta = ex.get("resposta", "")

        linhas.append(f'MESSAGE user """{pergunta}"""')
        linhas.append(f'MESSAGE assistant """{resposta}"""')

    return "\n".join(linhas)

    # 1. Modelo base
    linhas.append(f'FROM {modelo_base}\n')

    # 2. Parâmetros de geração
    linhas.append('PARAMETER temperature 0.7')
    linhas.append('PARAMETER num_predict 300\n')

    # 3. System prompt permanente
    linhas.append(f'SYSTEM """\n{system_prompt}\n"""\n')

    # 4. Exemplos de conversa (few-shot embutido no modelo)
    for ex in exemplos:
        linhas.append(f"MESSAGE user '{ex['pergunta']}'")
        linhas.append(f"MESSAGE assistant '{ex['reposta']}'\n")

    return '\n'.join(linhas)

# --- CONFIGURAÇÃO DO PROMPT DO PERSONAL TRAINER ---
SYSTEM_PROMPT = """
Você é o FitBot, um personal trainer virtual especializado em musculação e bem-estar.

Suas regras:
- Sempre responda em português brasileiro.
- Seja amigável, motivador, empático e objetivo.
- Use emojis relevantes para humanizar a conversa e incentivar o usuário.
- Se não souber a resposta técnica, oriente o usuário a procurar um profissional de educação física ou médico.
- Priorize a segurança: se o usuário relatar dor aguda, peça para ele parar o exercício.
- Foque em ajudar o usuário a entender a execução correta e a constância nos treinos.
- Sempre responda com pelo menos 3 partes: alerta, explicação e recomendação final.
- Sempre explique o motivo da orientação.
"""

# Gerando o conteúdo final
# Certifique-se de que a variável 'dataset' (da célula anterior) já foi executada
conteudo_modelfile = gerar_modelfile("llama3.2:1b", SYSTEM_PROMPT, dataset)

print("✨ Modelfile gerado com sucesso para o FitBot! \n")
print(conteudo_modelfile)
```

```python
# Escreve o conteúdo do Modelfile em um arquivo real no ambiente do Colab
with open("Modelfile", "w") as f:
    f.write(conteudo_modelfile)

# Comando para o Ollama criar o seu modelo personalizado
!ollama create fitbot -f Modelfile
```
